StandardScaler performs the task of Standardization. Usually a dataset contains variables that are different in scale. For e.g. an Employee dataset will contain AGE column with values on scale 20-70 and SALARY column with values on scale 10000-80000. As these two columns are different in scale, they are Standardized to have common scale while building machine learning model.

```python
backup_df
```

```python
scaler = StandardScaler()
backup_df[["Sales", "Quantity", "Profit"]] = scaler.fit_transform(backup_df[["Sales", "Quantity", "Profit"]])
```

```python
backup_df
```

Running actual K-Means model based on those 3 Columns

```python
kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
backup_df['Clusters'] = kmeans.fit_predict(backup_df[['Sales', 'Quantity', 'Profit']])
```

# Plotting the findings

```python
# visualize the clusters
plt.figure(figsize=(12,6), dpi = 100)
sns.scatterplot(data=backup_df, x="Sales", y="Profit", hue="Clusters")
plt.title("Customer Segmentation using K-Means Clustering")
plt.show()
```

# Choosing K Value Using ELBOW Method

```python
ssd = []

for k in range(2,10):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(backup_df[['Sales', 'Quantity', 'Profit']])
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,10),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
ssd
```

```python
pd.Series(ssd).diff()
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```
